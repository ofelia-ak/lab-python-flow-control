```python
products = ["t-shirt", "mug", "hat", "book", "keychain"]
inventory = {}
for product in products:
    quantity = int(input(f"Enter quantity available for {product}: "))
    inventory[product] = quantity
print(inventory)
```

```python
customer_orders = set()
for i in range(3):
    order = input("Enter the name of product for the order: ")
    if order in products:
        customer_orders.add(order)
print(customer_orders)
```

```python
customer_orders = set()

new_order = input("Enter the name of product for the order: ")
if new_order in products:
    customer_orders.add(new_order)
another_product = input("Do you want to add another product: ")
while another_product == "yes":
    new_order = input("Enter the name of product for the order: ")
    if new_order in products:
        customer_orders.add(new_order)
        print("Product added.")
    another_product = input("Do you want to add another product: ")
else:
    print("Order is complete")
   
print(customer_orders)
```

```python
print(inventory)
```

```python
for order in customer_orders:
    if order in inventory:
        inventory[order] -= 1
print(inventory)
```
